```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
import yfinance as yf
```

```python
df = yf.download("AAPL", start="2015-01-01", end="2025-01-01")
```

```python
df['Daily Return'] = df['Close'].pct_change()

plt.figure(figsize=(14, 6))
sns.lineplot(data=df['Daily Return'], color='blue')
plt.title('AAPL Daily Returns (20152025)', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['Daily Invest'] = df['Open'].pct_change()

plt.figure(figsize=(14, 6))
sns.lineplot(data=df['Daily Invest'], color='red')
plt.title('AAPL Daily Invest (20152025)', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Daily Invset')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['Daily High'] = df['High'].pct_change()

plt.figure(figsize=(14, 6))
sns.lineplot(data=df['Daily High'], color='yellow')
plt.title('AAPL Daily High (20152025)', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Daily High')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['Daily Low'] = df['Low'].pct_change()

plt.figure(figsize=(14, 6))
sns.lineplot(data=df['Daily Low'], color='green')
plt.title('AAPL Daily Low (20152025)', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Daily Low')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['Daily Volume'] = df['Volume'].pct_change()

plt.figure(figsize=(14, 6))
sns.lineplot(data=df['Daily Volume'], color='black')
plt.title('AAPL Daily Volume (20152025)', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Daily Volume')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['Rolling_STD'] = df['Close'].rolling(window=20).std()

plt.figure(figsize=(14, 6))
sns.lineplot(data=df['Rolling_STD'], color='Blue')
plt.title('AAPL Rolling_STD', fontsize=16)
plt.xlabel('Date')
plt.ylabel('STD')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['MA20'] = df['Close'].rolling(window=20).mean()

plt.figure(figsize=(14, 6))
sns.lineplot(data=df['MA20'], color='red')
plt.title('MA20', fontsize=16)
plt.xlabel('Date')
plt.ylabel('MA')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['Upper'] = df['MA20'] + 2*df['Rolling_STD']

plt.figure(figsize=(14, 6))
sns.lineplot(data=df['Upper'], color='green')
plt.title('Upper', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Upper')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['Lower'] = df['MA20'] - 2*df['Rolling_STD']

plt.figure(figsize=(14, 6))
sns.lineplot(data=df['Lower'], color='purple')
plt.title('Lower', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Lower')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'META']

data = yf.download(tickers, start="2022-01-01", end="2023-12-31")['Close']

daily_returns = data.pct_change()

corr_matrix = daily_returns.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Daily Returns (20222023)', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
rolling_corr = daily_returns['AAPL'].rolling(window=60).corr(daily_returns['MSFT'])
rolling_corr.plot(title='60-Day Rolling Correlation: AAPL vs MSFT', figsize=(12, 6))
```

Market reaction to global events (COVID-19, Fed rate changes)

| Date            | Event Description                     | Expected Market Reaction           |
| --------------- | ------------------------------------- | ---------------------------------- |
| **20 Feb 2020** | S\&P 500 peaked before COVID crash    | Start of major decline             |
| **16 Mar 2020** | Fed cut interest rate to 0%           | Temporary panic but stabilizing    |
| **23 Mar 2020** | Fed announced unlimited QE (stimulus) | Market bottomed and began recovery |
| **Apr 2020+**   | Stimulus measures took effect         | Sharp recovery phase               |


```python
ticker1 = 'SPY' 
df = yf.download(ticker1, start='2019-10-01', end='2020-09-30')

plt.figure(figsize=(14, 6))
plt.plot(df['Close'], label='S&P 500 Closing Price', color='blue')
plt.axvline(pd.to_datetime('2020-02-20'), color='red', linestyle='--', label='Market Peak (pre-COVID crash)')
plt.axvline(pd.to_datetime('2020-03-16'), color='orange', linestyle='--', label='Fed Emergency Rate Cut')
plt.axvline(pd.to_datetime('2020-03-23'), color='green', linestyle='--', label='Fed Unlimited QE Announcement')
plt.axvline(pd.to_datetime('2020-04-01'), color='purple', linestyle='--', label='Post-Stimulus Recovery')
plt.title('Market Reaction to COVID-19 Events - S&P 500', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['Volatility'] = df['Close'].pct_change().rolling(window=5).std()

plt.figure(figsize=(14, 5))
plt.plot(df['Volatility'], label='Rolling 5-Day Volatility', color='crimson')
plt.title('Volatility During COVID-19 Crisis')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
df = yf.download("SPY", start="2019-01-01", end="2024-12-31")

df['Daily Return'] = df['Close'].pct_change()

df = df.dropna()

top_gains = df.sort_values('Daily Return', ascending=False).head(10)
top_losses = df.sort_values('Daily Return').head(10)

print(" Top 10 Single-Day Gains:")
print(top_gains[['Daily Return']])

print("\n Top 10 Single-Day Losses:")
print(top_losses[['Daily Return']])
```

```python
plt.figure(figsize=(12, 5))
top_gains['Daily Return'].plot(kind='bar', color='green', title='Top 10 Single-Day Gains')
plt.ylabel('Daily Return')
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 5))
top_losses['Daily Return'].plot(kind='bar', color='red', title='Top 10 Single-Day Losses')
plt.ylabel('Daily Return')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
tickers = ['AAPL', 'MSFT']
data = yf.download(tickers, start='2019-01-01', end='2024-12-31')['Close']

daily_returns = data.pct_change()

rolling_corr = daily_returns['AAPL'].rolling(window=60).corr(daily_returns['MSFT'])

plt.figure(figsize=(14, 6))
plt.plot(rolling_corr, label='60-Day Rolling Correlation (AAPL vs MSFT)', color='blue')
plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.title('Rolling Correlation Between AAPL and MSFT (20192024)', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Correlation')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
